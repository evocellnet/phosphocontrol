```python
comparison_label = ["PP", "PN", "NN"]
is_nan = np.zeros((len(psites), 3), dtype = bool)

for i, psite in enumerate(psites):
    res = int(psite.split("_")[1])
    
    for j, label in enumerate(comparison_label):
        try:
            all_comparisons = np.load("data/server/output_cd/" + psite + "_" + label + ".npy", allow_pickle = True)
            for comp in all_comparisons:
                x = len([array for array in comp if np.sum(np.isnan(array)) == 0])
                if x == 0:
                    is_nan[i, j] = True
        except:
            is_nan[i, j] = True
```

# Defining functions to calculate mean strain

```python
def geom_mean(x):
    return np.exp(np.mean(np.log(x)))

def mean_l3(all_comparisons, res, groups, random = False, seed = 0):
    fsite_temp = []
    bground_temp = []
    psite_temp = []

    for group in groups:
        fsite_temp.append([])
        
    for c, comp in enumerate(all_comparisons):            
        if type(comp) == tuple:
            interval = np.sum([x == comp[2] for x in [res-2, res-1, res, res+1, res+2]], axis = 0, dtype = bool)
            other = ~interval    

            if random:
                np.random.seed(seed)
                size = len(comp[1])
                l3 = comp[1][np.random.choice(size, size, replace = False)]
            else:
                l3 = comp[1]

            for g, group in enumerate(groups):
                data_group = data_prot[data_prot["site_idx"] == group]
                res_group = list(data_group["residue"])
                
                res_group = np.sum([x == comp[2] for x in res_group], axis = 0, dtype = bool)
                other = ~(~other + res_group)
                
                if np.sum(res_group) != 0:
                    fsite_temp[g].append(geom_mean(l3[res_group]))
    
            psite_temp.append(geom_mean(l3[interval]))
            bground_temp.append(geom_mean(l3[other]))

    return np.nanmean(psite_temp), np.nanmean(bground_temp), [np.nanmean(g) for g in fsite_temp]

def prot_l3(all_comparisons, random = False, seed = 0):
    prot_temp = []
        
    for c, comp in enumerate(all_comparisons):            
        if type(comp) == tuple:
            if random:
                np.random.seed(seed)
                size = len(comp[1])
                l3 = comp[1][np.random.choice(size, size, replace = False)]
            else:
                l3 = comp[1]

            prot_temp.append(geom_mean(l3))

    return np.nanmean(prot_temp)
```

# Strain analyses - Enrichment test

```python
valid_fsite = [[], [], []]
str_fsite = [[], [], []]
str_both = [[], [], []]

for j in range(3):
    for g in valid_groups[j]:
        data_group = annot[annot["site_idx"] == g]
        valid_fsite[j].append(data_group.iloc[0, 5])

    for g in str_fsite_g[j]:
        data_group = annot[annot["site_idx"] == g]
        str_fsite[j].append(data_group.iloc[0, 5])

    for g in str_both_g[j]:
        data_group = annot[annot["site_idx"] == g]
        str_both[j].append(data_group.iloc[0, 5])
```

```python
# survival function (k, M, n, N)
# k success in sampled
# N size of dataset
# n success in dataset
# M size of dataset

from scipy.stats import hypergeom

def trim_uniprot(vec):
    res = []
    
    for x in vec:
        x = x.split(":")[1:]
        t = ""
        for i in x:
            t += i + ":"
        res.append(t)
    return res

dataset_hg = trim_uniprot(valid_fsite[1])
sampled_hg = trim_uniprot(str_both[1])

M = len(dataset_hg)
N = len(sampled_hg)
nrep = len(np.unique(dataset_hg))
for uniq in np.unique(dataset_hg):
    n = dataset_hg.count(uniq)
    k = sampled_hg.count(uniq)
    pval = hypergeom.sf(k, M = M, n = n, N = N) < .05 / nrep
    if pval:
        print(uniq, pval, dataset_hg.count(uniq), sampled_hg.count(uniq))
        print(hypergeom.sf(k, M = M, n = n, N = N))
```

# Shortest distance between P-site and F-site

```python
def shortest_dist_pfsite(pres, fres, psite):
    accessions = protN_accession[psite] + protP_accession[psite]

    shortest = []
    for accession in accessions:
        file = "data/server/output_pdist/" + accession + "_pdist.npy"
        if not os.path.exists(file):
            continue
        
        vector = np.load(file)
        nres = (1 + 8*vector.shape[0])**.5/2 - 1/2
        nres = int(nres)
        
        idx = vector[:nres].tolist()
        pdist = squareform(vector[nres:])
        
        if pres in idx:
            fres_in = False
            min_dist = np.inf
            i = idx.index(pres)
            
            for fres_j in fres:
                if fres_j in idx:
                    fres_in = True
                    j = idx.index(fres_j)
                    
                    if pdist[i, j] < min_dist:
                        min_dist = pdist[i, j]

            if fres_in:
                shortest.append(min_dist)

    if len(shortest):
        return np.median(shortest), np.min(shortest)
    return [np.nan]
    
```

```python
def mean_overlap(all_comparisons, res, groups, random = False, seed = 0):
    fsite_temp = []
    psite_temp = []

    for group in groups:
        fsite_temp.append([])
        
    for c, comp in enumerate(all_comparisons):
        if type(comp) == tuple:
            if random:
                np.random.seed(seed)
                size = len(comp[1])
                l3 = comp[1][np.random.choice(size, size, replace = False)]
            else:
                l3 = comp[1]

            N = np.argmin(comp[0]) + 10

            min_cd = set(np.argsort(l3)[::-1][:N])
            interval = set(np.where(np.sum([x == comp[2] for x in [res-2, res-1, res, res+1, res+2]], axis = 0, dtype = bool))[0])
            
            if len(interval) > 0:
                psite_temp.append(len(interval.intersection(min_cd)) / len(interval))
            
            for g, group in enumerate(groups):
                data_group = data_prot[data_prot["site_idx"] == group]
                res_group = list(data_group["residue"]) 
                res_group = set(np.where(np.sum([x == comp[2] for x in res_group], axis = 0, dtype = bool))[0])

                if len(res_group) > 0:
                    fsite_temp[g].append(len(res_group.intersection(min_cd)) / len(res_group))

    return np.nanmean(psite_temp), [np.nanmean(g) for g in fsite_temp]
```
