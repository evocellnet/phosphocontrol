```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from functools import reduce
from copy import deepcopy

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
# Need to add:
# - local changes in fluctuations
# - local changes around the phosphosite
# - changes in mutual information (RV2)
# - changes in mutual information (MI)  
# - strain 
```

```python
##########################
# Conformational changes #
##########################

# Backbone comparison
rmsd_df = pd.read_csv("backbone_comparison/rmsd_merged_df.csv")

####################
# Dynamics changes #
####################

# RMSIPs
rmsip_df = pd.read_csv("global_dynamics_overview/rmsips_df_nodups.csv",index_col=0)

# Fluctuation correlations
flucts_corr_df = pd.read_csv("global_dynamics_overview/fluctuations_corr_df_nodups.csv")

# Overall fluctuations
flucts_df = pd.read_csv("final_dynamics_overview/fluctuations_df_nodups.csv")
# Rename columns
col_mapper = {'median_structure_phospho':'median_flucts_phospho',
              'median_structure_nonphospho':'median_flucts_nonphospho',
             'std_structure_phospho':'std_flucts_phospho',
             'std_structure_nonphospho':'std_flucts_nonphospho'}
flucts_df.rename(col_mapper,axis=1,inplace=True)

flucts_df['fluctuations_diff'] = flucts_df['median_flucts_phospho'] - flucts_df['median_flucts_nonphospho']

# Local fluctuations
joint_local_dynamics_df = pd.read_csv("peak_cutoff_height/joint_local_dynamics_df.csv")

# Phosphosite dynamics
psite_dynamics_df = pd.read_csv("psite_local_dynamics/psite_dynamics_df.csv",index_col=0)

psite_dynamics_df = psite_dynamics_df.rename({"-log10_adj_pval":"-log10_adj_pval_psite","diffs":"psite_fluctuation_diffs"},axis=1)
psite_dynamics_df = psite_dynamics_df[["phosphosite","-log10_adj_pval_psite","psite_fluctuation_diffs"]]

# Linear mutual information
lmi_df = pd.read_csv("final_dynamics_overview/lmi_df_nodups.csv",index_col=0)
```

```python
extensions_df = pd.read_csv("../results/strain_results/extensions_mean_psite_vs_others.csv")
perc_in_manifold_df = pd.read_csv("../results/strain_results/percentage_in_manifold_psite.csv")

# "../results/strain_results/extensions_uniprot_psite_actsite_other.csv"
# "../results/strain_results/percentage_in_manifold_actsite.csv"
# "../results/strain_results/percentage_in_manifold_psite.csv"

extensions_df = extensions_df[["Psite","PN_other","PN_psite_+-_2"]].dropna()
extensions_df = extensions_df.rename({'Psite':'phosphosite','PN_other':'lambda3_general','PN_psite_+-_2':'lambda3_psite'},axis=1)

perc_in_manifold_df = perc_in_manifold_df[["Psite","PN_psite"]]
perc_in_manifold_df = perc_in_manifold_df.rename({'Psite':'phosphosite','PN_psite':'perc_in_manifold'},axis=1)
```

```python
merged_df = rmsd_df.merge(rmsip_df,on="phosphosite").merge(lmi_df,on="phosphosite").merge(flucts_corr_df,on="phosphosite").merge(flucts_df,on="phosphosite").merge(joint_local_dynamics_df, on="phosphosite").merge(psite_dynamics_df, on="phosphosite")
```

```python
#df_list = [rmsd_df, flucts_corr_df, flucts_df, joint_local_dynamics_df, lmi_df]
#df_list = [rmsd_df, rmsip_df, flucts_corr_df, flucts_df, joint_local_dynamics_df, psite_dynamics_df, lmi_df, extensions_df]
#df_list = [rmsd_df, rmsip_df, flucts_corr_df, flucts_df, joint_local_dynamics_df, lmi_df]

df_list = [rmsd_df, rmsip_df, flucts_corr_df, flucts_df, joint_local_dynamics_df, lmi_df,extensions_df]

merged_df = reduce(lambda x,y: x.merge(y,on='phosphosite'), df_list)
```

```python

test_df.columns[test_df.isna().any()].tolist()
# Double check where the NAs are
test_df = test_df.fillna(0)
```

```python
# Would be cool if we can add functional scores?
# Need also to reflect NAs (no functional scores known)
# sns.clustermap(test_df, cmap='vlag',row_colors=colors, method="average",
#                metric="correlation",z_score=1,center=0)
sns.clustermap(test_df, cmap='coolwarm',row_colors=colors, method="average",
                metric="correlation",z_score=1,center=0)
```

```python
test_df_transformed = MinMaxScaler().fit_transform(test_df)
test_df_transformed = pd.DataFrame(test_df_transformed, columns=test_df.columns)
```

```python
sns.clustermap(test_df_transformed, cmap='viridis',row_colors=colors, method="average",
                metric="correlation")
```

```python
pca = PCA(n_components=2)
test_df_reduced = pca.fit_transform(test_df_transformed)
```

```python
plt.scatter(test_df_reduced[::,0],test_df_reduced[::,1],alpha=0.4,color=colors)
plt.xlabel(f"PC 1 ({100*pca.explained_variance_ratio_[0]:.2f}%)")
plt.ylabel(f"PC 2 ({100*pca.explained_variance_ratio_[1]:.2f}%)")
sns.despine()
```

```python
plt.scatter(test_df_reduced[::,0],test_df_reduced[::,1],alpha=0.4,color=colors)
plt.xlabel(f"PC 1 ({100*pca.explained_variance_ratio_[0]:.2f}%)")
plt.ylabel(f"PC 2 ({100*pca.explained_variance_ratio_[1]:.2f}%)")
plt.xlim(-0.3,0.5)
sns.despine()
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.ylabel("Explained variance ratio")
plt.xlabel('Component')
sns.despine()
plt.ylim(0, 0.8)
```

```python
weights = pca.components_
```

```python
len(test_df.columns)
```

```python
feature_weights = []
for idx, val in enumerate(weights[0]):
    #print(idx, test_df.columns[idx], val)
    print(test_df.columns[idx],val)
    feature_weights.append([test_df.columns[idx], val])
feature_weights = pd.DataFrame(feature_weights,columns=["feature","weight"])
```

```python
sns.set_context("talk")
sns.barplot(feature_weights.sort_values(by="weight"), y="feature", x="weight",color='black')
sns.despine()
```

```python
from hdbscan import HDBSCAN
```

```python
clusterer = HDBSCAN(metric="euclidean")
clusterer.fit(test_df_reduced)
labels = clusterer.labels_
```

```python
test_df_reduced.shape
```

```python
plt.scatter(test_df_reduced[::,0],test_df_reduced[::,1],alpha=0.4,c=labels)
plt.xlabel(f"PC 1 ({100*pca.explained_variance_ratio_[0]:.2f}%)")
plt.ylabel(f"PC 2 ({100*pca.explained_variance_ratio_[1]:.2f}%)")
#plt.xlim(-0.3,0.3)
sns.despine()
```
