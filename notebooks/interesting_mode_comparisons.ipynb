Notebook to inspect some possible interesting examples of changes in normal modes of motion

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
out_path = Path("mode_comparison_examples")
out_path.mkdir(exist_ok=True)
```

```python
def rename_index_and_cols(df):
    column_mapper = {}
    index_mapper = {}
    
    for column in df.columns:
        column_name = column.replace("b","")
        column_mapper[column] = column_name
        
    for row in df.index:
        row_name = row.replace("a","")
        index_mapper[row] = row_name

    df = df.rename(column_mapper, axis=1)
    df = df.rename(index_mapper, axis=0)
    return df
```

```python
def rename_cols(df):
    column_mapper = {}
    
    for column in df.columns:
        column_name = column.replace("V","")
        column_mapper[column] = column_name
        
    df = df.rename(column_mapper, axis=1)
    return df
```

# Pyruvate dehydrogenase E1 component subunit alpha
RMSIP between groups=0.990674

```python
base_path = Path("/local/home/mcorrea/Documents/controlphospho/results/cool_mode_examples/")
```

```python
pdh_path = base_path / "P08559_293"
between_pdh = pd.read_csv(pdh_path / "between_median_overlap.csv",index_col=0)
between_pdh = rename_cols(between_pdh)

within_np_pdh = pd.read_csv(pdh_path / "within_np_overlap.csv",index_col=0)
within_np_pdh = rename_cols(within_np_pdh)

within_p_pdh = pd.read_csv(pdh_path / "within_p_overlap.csv",index_col=0)
within_p_pdh = rename_cols(within_p_pdh)
```

```python
sns.set_context('talk')

fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(24, 9))


sns.heatmap(between_pdh, vmin=0,vmax=1,cmap='flare_r',cbar_kws={'label': 'Mode RMSIP'},ax=ax0,cbar=False)
ax0.set_ylabel("Non-phosphorylated")
ax0.set_xlabel("Phosphorylated")
ax0.set_title("Phosphorylated vs non-phosphorylated")

sns.heatmap(within_np_pdh, vmin=0,vmax=1,cmap='flare_r',cbar_kws={'label': 'Mode RMSIP'},ax=ax1,cbar=False)
ax1.set_ylabel("Non-phosphorylated")
ax1.set_xlabel("Non-phosphorylated")
ax1.set_title("Within non-phosphorylated proteins")


sns.heatmap(within_p_pdh, vmin=0,vmax=1,cmap='flare_r',cbar_kws={'label': 'Mode RMSIP'},ax=ax2)
ax2.set_ylabel("Phosphorylated")
ax2.set_xlabel("Phosphorylated")
ax2.set_title("Within phosphorylated proteins")

plt.suptitle("Pyruvate dehydrogenase E1 component subunit alpha")

plt.savefig(pdh_path / "pdh_rmsip_mode_comparison.pdf",dpi=120,bbox_inches='tight')
```
